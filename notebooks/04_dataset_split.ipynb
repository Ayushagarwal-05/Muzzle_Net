```python
from pathlib import Path
import random
import shutil

dataset_path = Path("../datasets/identity_dataset")

all_classes = sorted([
    folder.name
    for folder in dataset_path.iterdir()
    if folder.is_dir()
])

random.seed(42)
random.shuffle(all_classes)

total = len(all_classes)

train_split = int(total * 0.7)
val_split = int(total * 0.15)

train_classes = all_classes[:train_split]
val_classes = all_classes[train_split:train_split+val_split]
test_classes = all_classes[train_split+val_split:]

print("Train:", len(train_classes))
print("Validation:", len(val_classes))
print("Test:", len(test_classes))
```
