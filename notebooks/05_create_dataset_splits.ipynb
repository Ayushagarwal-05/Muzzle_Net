```python
from pathlib import Path
import shutil

# SOURCE DATASET
source = Path("../datasets/identity_dataset")

# DESTINATION
destination = Path("../datasets/identity_split")

# TRAIN IDS
train_ids = (
    list(range(1,11)) +
    list(range(12,21)) +
    list(range(22,73)) +
    list(range(74,87))
)

# VALIDATION IDS
val_ids = list(range(87,105))

# TEST IDS
test_ids = (
    list(range(105,121)) +
    list(range(122,125))
)

def copy_folders(id_list, split_name):

    split_path = destination / split_name
    split_path.mkdir(parents=True, exist_ok=True)

    for idx in id_list:

        folder_name = f"cattle_id_{idx:03d}"

        src = source / folder_name
        dst = split_path / folder_name

        if src.exists():

            shutil.copytree(
                src,
                dst,
                dirs_exist_ok=True
            )

            print(f"Copied {folder_name} → {split_name}")

        else:
            print(f"Missing: {folder_name}")


copy_folders(train_ids, "train")
copy_folders(val_ids, "val")
copy_folders(test_ids, "test")

print("\nDataset split completed.")
```
