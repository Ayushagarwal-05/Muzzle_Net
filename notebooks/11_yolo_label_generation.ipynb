```python
from pathlib import Path
import shutil
import random
```

```python
source = Path(
    "../datasets/dataset/Original/OriginalMaster"
)

target = Path(
    "../datasets/yolo_muzzle"
)

for folder in [
    target/"images/train",
    target/"images/val",
    target/"labels/train",
    target/"labels/val"
]:
    folder.mkdir(
        parents=True,
        exist_ok=True
    )
```

```python
images = sorted(
    list(source.glob("*.jpg"))
)

random.seed(42)
random.shuffle(images)

split = int(
    len(images)*0.8
)

train_images = images[:split]
val_images = images[split:]

print(
    len(train_images),
    len(val_images)
)
```

```python
def create_yolo_label():

    x_center = 0.50
    y_center = 0.575

    width = 0.50
    height = 0.45

    return (
        f"0 "
        f"{x_center} "
        f"{y_center} "
        f"{width} "
        f"{height}"
    )
```

```python
for img_path in train_images:

    shutil.copy(
        img_path,
        target/"images/train"/img_path.name
    )

    label_path = (
        target/"labels/train"/
        img_path.with_suffix(".txt").name
    )

    with open(
        label_path,
        "w"
    ) as f:

        f.write(
            create_yolo_label()
        )
```

```python
for img_path in val_images:

    shutil.copy(
        img_path,
        target/"images/val"/img_path.name
    )

    label_path = (
        target/"labels/val"/
        img_path.with_suffix(".txt").name
    )

    with open(
        label_path,
        "w"
    ) as f:

        f.write(
            create_yolo_label()
        )
```

```python
yaml_text = """
path: ../datasets/yolo_muzzle

train: images/train
val: images/val

names:
  0: muzzle
"""

with open(
    target/"dataset.yaml",
    "w"
) as f:

    f.write(
        yaml_text
    )

print(
    "YOLO dataset ready."
)
```
