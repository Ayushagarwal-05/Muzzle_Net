```python
from ultralytics import YOLO
```

```python
model = YOLO(
    "../notebooks/runs/models/muzzle_detector/weights/best.pt"
)

print("Model loaded.")
```

```python
from pathlib import Path
import random

images = list(
    Path(
        "../datasets/dataset/Original/OriginalMaster"
    ).glob("*.jpg")
)

test_image = random.choice(images)

print(test_image)
```

```python
results = model.predict(
    source=str(test_image),
    conf=0.25,
    save=False
)
```

```python
from pathlib import Path
import random
import shutil

source = Path("../datasets/dataset/Original/OriginalMaster")
target = Path("../datasets/manual_labeling")

target.mkdir(exist_ok=True)

images = list(source.glob("*.jpg"))

random.seed(42)
selected = random.sample(images, 200)

for img in selected:
    shutil.copy(img, target / img.name)

print("Selected:", len(selected))
```
