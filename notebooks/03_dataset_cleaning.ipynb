```python
from pathlib import Path
import shutil

dataset_path = Path("../datasets/identity_dataset")

MIN_IMAGES = 4

removed = []

for folder in dataset_path.iterdir():
    
    if folder.is_dir():
        count = len(list(folder.glob("*")))
        
        if count < MIN_IMAGES:
            removed.append((folder.name, count))

print("Classes to remove:")
print(removed)
```
